```python
#Import Library
import matplotlib.pyplot as plt
#%matplotlib inline

from skimage import data
from skimage.io import imread
from skimage.color import rgb2gray 
from skimage.util import invert

import numpy as np
import cv2
```

```python
#percobaan 1 : crop image
# Load gambar dari local drive
akuImage = cv2.imread("aku.jpeg")
matchaImage = cv2.imread("matcha.jpg")
# Cropping gambar
akuCropped = akuImage.copy()
akuCropped = akuCropped[400:800,128:500]

matchaCropped = matchaImage.copy()
matchaCropped = matchaCropped[400:800,128:500]

print('Aku Ori Shape : ',akuImage.shape)
print('Aku Crop Shape : ',akuCropped.shape)

print('Camera Ori Shape : ',matchaImage.shape)
print('Camera Crop Shape : ',matchaCropped.shape)
# Membuat figure subplot
fig, axes = plt.subplots(2, 2, figsize=(8, 8))
ax = axes.ravel()
# Plot Gambar pertama dengan cv2.imshow
ax[0].imshow(cv2.cvtColor(akuImage, cv2.COLOR_BGR2RGB))
ax[0].set_title("Citra Input 1")

ax[1].imshow(cv2.cvtColor(matchaImage, cv2.COLOR_BGR2RGB))
ax[1].set_title('Citra Input 2')
# Plot gambar ketiga dengan cv2.imshow
akuCropped_resized = cv2.resize(cv2.cvtColor(akuCropped, cv2.COLOR_BGR2RGB), (320, 320))
ax[2].imshow(akuCropped_resized)
ax[2].set_title("Citra Output 1")

# Plot gambar keempat dengan cv2.imshow
matchaCropped_resized = cv2.resize(cv2.cvtColor(matchaCropped, cv2.COLOR_BGR2RGB), (800, 800))

ax[3].imshow(matchaCropped_resized)
ax[3].set_title('Citra Output 2')

plt.show()
```

```python
#Fungsi invert digunakan untuk membuat gambar negatif dari citra input
inv = invert(akuCropped)
#mencetak bentuk (shape) dari citra input dan citra output
print('Shape Input : ', akuCropped.shape)
print('Shape Output : ',inv.shape)
# Membuat figure subplot
fig, axes = plt.subplots(2, 2, figsize=(8, 8))
ax = axes.ravel()
# Plot Gambar Input dengan cv2.imshow
ax[0].imshow(cv2.cvtColor(akuImage, cv2.COLOR_BGR2RGB))
ax[0].set_title("Citra Input")

# Plot Histogram Input
ax[1].hist(akuCropped.ravel(), bins=256)
ax[1].set_title('Histogram Input')

# Plot Gambar Output Negatif
ax[2].imshow(inv)
ax[2].set_title('Citra Output (Inverted Image)')

# Plot Histogram Output Negatif
ax[3].hist(inv.ravel(), bins=256)
ax[3].set_title('Histogram Output')
```

```python
#membuat salinan dari citra dan mengubah tipenya menjadi float 
copymatcha = matchaCropped.copy().astype(float)
#menentukan kecerahan
brightness = 100
matcha_bright = np.clip(copymatcha + brightness, 0, 255).astype(np.uint8)

# Membuat figure subplot  
fig, axes = plt.subplots(2, 2, figsize=(8, 8))
ax = axes.ravel()
# Plot Gambar Input dengan cv2.imshow
ax[0].imshow(cv2.cvtColor(matchaCropped, cv2.COLOR_BGR2RGB))
ax[0].set_title("Citra Input")
# Plot Histogram Input
ax[1].hist(matchaCropped.ravel(), bins=256)
ax[1].set_title('Histogram Input')
# Plot Gambar Output output
ax[2].imshow(cv2.cvtColor(matcha_bright, cv2.COLOR_BGR2RGB))
ax[2].set_title(f"Citra Output (Brightness={brightness})")
# Plot Histogram output
ax[3].hist(matcha_bright.ravel(), bins=256, range=(0, 255))
ax[3].set_title('Histogram Output')

plt.show()
```

```python
#import library
import numpy as np
import cv2
import matplotlib.pyplot as plt
```

```python
#load gambar
img = cv2.imread("aku.jpeg")
```

```python
#mendapatkan/define resolusi dan tipe gambar
img_height=img.shape[0]#pengambilan nilai tinggi citra 
img_width = img.shape[1]#pengambilan nilai lebar citra
img_channel= img.shape[2]#pengambilan jumlah channel citra 
img_type= img.dtype #pengambilan tipe data (data type) dari citra
```

```python
#percobaan pertama kita buat brightness untuk gambar grayscale 
#membuat variable img_brightness untuk menampung
img_brightness = np.zeros(img.shape, dtype = np.uint8)
```

```python
#membuat fungsi penambahan brightness dengan nilai yang menjadi parameter 
def brighter(nilai):
    for y in range(0, img_height): #perulangan for untuk mengakses setiap baris (y) pada citra digital dengan rentang 0 sampai img_height.
        for x in range (0, img_width):#perulangan for untuk mengakses setiap piksel pada baris (y) tersebut, dengan rentang 0 sampai img_width.
            #nilai intensitas warna merah, hijau, dan biru (red, green, dan blue) dari piksel yang sedang diakses 
            red= img[y][x][0]
            green= img[y][x][1]
            blue= img[y][x][2]
            #nilai keabuan (gray) dihitung dengan mengambil rata-rata nilai intensitas warna merah, hijau, dan biru dari piksel 
            gray= (int(red)+ int(green)+int(blue))/3
            #nilai keabuan (gray) ditingkatkan sebanyak "nilai" yang diberikan oleh parameter fungsi brighter
            gray += nilai
            #jika nilai keabuan (gray) lebih besar dari 225 maka nilai keabuan diatur ke nilai maksimum yang dapat direpresentasikan oleh tipe data uint8, yaitu 255.
            if gray > 225:
                gray=255
            #nilai keabuan (gray) lebih kecil dari 0. Jika ya, maka nilai keabuan diatur ke 0
            if gray<0:
                gray=0
            #piksel yang sedang diakses pada baris y dan kolom x.
            img_brightness[y][x]= (gray, gray, gray)
```

```python
#menampilkan gambar dgn parameter kecerahan -100
brighter(-100)
plt.imshow(img_brightness)
plt.title("Brightness -100")
plt.show()
#menampilkan gambar dgn parameter kecerahan 25
brighter(25)
plt.imshow(img_brightness)
plt.title ("Brightness 100")
plt.show()
```

```python
#contrass autoleave
img_autocontrass = np.zeros (img.shape, dtype=np.uint8)
```

```python
#melakukan kontras otomatis pada gambar
def autocontrass ():
  #melakukan inisialisasi variabel  nilai maksimum dan minimum dari intensitas warna
  xmax = 255
  xmin= 0
  d = 0

  for y in range (0,img_height):#perulangan for untuk mengakses setiap baris (y) pada citra digital dengan rentang 0 sampai img_height.
    for x in range (0, img_width):#perulangan for untuk mengakses setiap piksel pada baris (y) tersebut, dengan rentang 0 sampai img_width.
  
      #nilai intensitas warna merah, hijau, dan biru (red, green, dan blue) dari piksel yang sedang diakses 
      red = img [y][x][0]
      green = img[y][x][1]
      blue = img [y][x][2]
      #nilai keabuan (gray) dihitung dengan mengambil rata-rata nilai intensitas warna merah, hijau, dan biru dari piksel 
      gray= (int (red)+int (green)+int (blue))/3
      #Nilai xmax akan diupdate dengan nilai gray jika gray kurang dari xmax
      if gray < xmax :
        xmax = gray
      #nilai xmin akan diupdate dengan nilai gray jika gray lebih besar dari xmin
      if gray >xmin:
        xmin=gray
  #variabel d akan diupdate dengan selisih antara nilai xmin dan xmax
  d = xmin-xmax
  for y in range (0,img_height):#perulangan for untuk mengakses setiap baris (y) pada citra digital dengan rentang 0 sampai img_height.
    for x in range (0, img_width):#perulangan for untuk mengakses setiap piksel pada baris (y) tersebut, dengan rentang 0 sampai img_width.
      #nilai intensitas warna merah, hijau, dan biru (red, green, dan blue) dari piksel yang sedang diakses 
      red = img [y][x][0]
      green = img[y][x][1]
      blue = img [y][x][2]
      #nilai keabuan (gray) dihitung dengan mengambil rata-rata nilai intensitas warna merah, hijau, dan biru dari piksel 
      gray= (int (red)+int (green)+int (blue))/3
      gray = int(float(255/d)*(gray-xmax)) #rumusnya
      #piksel yang sedang diakses pada baris y dan kolom x
      img_autocontrass [y][x]= (gray,gray,gray)
```
